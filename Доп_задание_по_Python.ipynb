```python
# Задание 1: Напишите программу, которая найдет все такие числа, которые делятся на 7, но не делятся на 5, от 2000 до 3200 (включая оба).
# Полученные числа должны быть напечатаны в последовательности, разделенной запятыми, в одной строке.
l=[]
for k in range(2000, 3201):
   if k%7==0 and k%5!=0:
       l.append(str(k))
print(l)
```

```python
# Задание 2: Напишите программу, которая может вычислить факториалы заданных чисел. Результаты должны быть напечатаны в последовательности, разделенной запятыми, в одной строке.
#Факториалом числа называют произведение всех натуральных чисел до него включительно. Например, факториал числа 5 равен произведению 1 * 2 * 3 * 4 * 5 = 120.

# Формула нахождения факториала:

# n! = 1 * 2 * … * n,

# где n – это число, а n! – факториал этого числа.

# Формулу можно представить в таком виде:

# n! = 1 * … * (n-2) * (n-1) * n,

# т.е. каждый предыдущий множитель меньше на единицу, чем последующий.


from math import factorial
l = [9, 15, 3, 1, 20]
for k in range(len(l)):
  if l[k] >= 0:
    print("\nДля числа ", l[k], "факториал равен",factorial(l[k]))
  else:
    print(l[k], "не натуральное число")
```

```python
# Задание 3: Напишите программу, которая принимает 2 числа X, Y в качестве входных данных с клавиатуры и генерирует двумерный массив. Значение элемента в i-й строке и j-м столбце массива должно быть i*j. (нумерация с нуля)

# Допустим для: i=3, j=5. Результат будет [ [0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8] ]


def matrix_k (x, y):
    matrix = [[0 for j in range(y)] for i in range(x)]
    for i in range(x):
        for j in range(y):
            matrix[i][j] = j * i
    print("матрица", *matrix, sep='\n') #print после окончания работы цикла

x = int(input())
y = int(input())
matrix_k (x, y)
```

```python
# Задание 4: Веб-сайт требует, чтобы пользователи вводили имя пользователя и пароль для регистрации. Напишите программу для проверки правильности ввода пароля пользователями.
# Ниже приведены критерии проверки пароля:

# Как минимум 1 буква между [a-z] По крайней мере, 1 число между [0–9] По крайней мере, 1 буква между [A–Z] По крайней мере, 1 символ из [$#@] Минимальная длина пароля транзакции: 6 Максимальная длина пароля транзакции: 12


import re
def parol():
  print('Введите пароль.')
  password = input()
  if len(password)<6:
    print('Пароль должен быть более 6 символов')
  elif len(password)>12:
    print('Пароль должен быть менее 12 символов')
  elif not re.search("[$#@]", password):
    print('Пароль должен содержать один из знаков $#@')
  elif not re.search("[A–Z]", password):
    print('Пароль должен содержать одину букву из A–Z')
  elif not re.search("[0-9]", password):
    print('Пароль должен содержать одино число из 0-9')
  elif not re.search("[a-z]", password):
    print('Пароль должен содержать одину букву из a-z')
  else:
    print('Пароль корректный')

parol()

```
