### 7. Для первых 30 строк постройте графики областей «произведенных» и «дефектных» деалей. (используйте .loc)

```python
df_new = df3.loc[:30,['defective','produced']]
df_new.plot.area(alpha=0.4, color = ['#ff7f0e','#1f77b4'])
```

```python
# ОЖИДАЕМЫЙ РЕЗУЛЬТАТ
```
